```
import numpy as np

# load data
import os
from google.colab import drive
drive.mount('/content/drive')
filedir = './drive/My Drive/Final/mfcc_data'
X = np.zeros((300, 12 * 63))
Y = np.zeros(300).astype(np.int32)
filelist = os.listdir(filedir)
```

```

for i in range(300):
  file = filelist[i]
  with open(filedir + '/' + file, 'rb') as f:
    data = np.load(f).flatten()
  data_size = min(len(data), 12 * 63)
  X[i,:data_size] = data[:data_size]
  if (file[0] == '0'):
    Y[i] = 0
  else:
    Y[i] = 1
```

```
with open(filedir + '/' + 'X', 'wb') as f:
  np.save(f, X)
with open(filedir + '/' + 'Y', 'wb') as f:
  np.save(f, Y)
```
